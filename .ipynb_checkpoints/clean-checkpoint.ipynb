## Data Cleaning 

To Do List: 
- Clean the NaN
- Clean irrelevant columns

```python
# dropping unnecessary columns 
df = df.drop(
    columns=[
        "Longitude",
        "Latitude",
        "Inspection Business Name",
        "Business_ID", 
        "Inspection_Serial_Num",
        "Violation_Record_ID",
        "Phone",
        "City",
        "Address",
        "Program Identifier"
    ]
)
```

```python
df.info()
```

```python
df.head(5)
```

```python
# # setting the index of the dataset 
# df.set_index(["Name"], inplace = True, 
#                     append = True, drop = True) 
# df.reset_index(drop = True)
# df.head(5)
```

```python
# drop rows with NaN in column of Name
df = df[pd.notnull(df['Name'])]
df.head(10)
df.reset_index(drop = True)
```

```python
# group by restaurant and get the sum of inspection score
df["UNAME"] = df["Zip Code"].map(str) + ' ' + df["Name"]
df_gb_restaurant = df.groupby(['UNAME', 'Zip Code'])['Inspection Score'].median().to_frame()
df_gb_restaurant
```

```python
df.drop(columns = ['UNAME']) # house keeping 
```

```python
# creating a rank column and passing the returned rank series 
df_gb_restaurant["Rank"] = df_gb_restaurant["Inspection Score"].rank() 

# sorting w.r.t name column 
df_gb_restaurant.sort_values("Rank", ascending= False, inplace = True) 
df_gb_restaurant
```
